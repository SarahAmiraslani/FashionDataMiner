```python
import pandas as pd
```

### import higg and material facts data

```python
material_info = pd.read_csv('mat_comments.csv')
higg = pd.read_csv('higg_rescaled.csv')
```

```python
higg.head()
```

```python
#merge
df = higg.merge(material_info, left_on='backend_higg', right_on='Material Name',how='outer')
df.drop(['Material Name', 'Sources'],axis=1, inplace=True)
```

```python
df.drop(127,axis=0, inplace=True)
```

### export data

```python
df.to_csv('higg_facts.csv', index=False)
```
