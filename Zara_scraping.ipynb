```python
#def get_zara_information(item_link):
zara=[]
import time
for i in item_link.keys():
    zara_result={}
    driver.get(item_link[i])
    more_buttons = driver.find_elements_by_class_name("_product-composition")
    for x in range(len(more_buttons)):
        if more_buttons[x].is_displayed():
            driver.execute_script("arguments[0].click();", more_buttons[x])
            time.sleep(0.25)
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'lxml')
    time.sleep(0.25)
    try:
        a=soup.find('ul', class_='list-composition')
        section=a.find_all('li',class_='section')#should modify to add try in case some clothes are check in store
        t={}
        for j in section:
            title=j.find('p',class_='title-section').get_text() #different parts of the item
            body=j.find_all('p',class_='zonasPrenda')#related material
            t[title]=[]
            for m in body:
                if m.get_text()!='':
                    t[title].append(m.get_text())

        #size
        sizes=soup.find_all('div',class_='size-name')
        size=[]
        for j in sizes:
            size.append(j.get_text())
        #image
        images=soup.find_all('img',class_='image-big _img-zoom _main-image')
        image=[]
        for j in images:
            image.append(j.get('src'))    


        #'Name', 'Material', 'Color', 'URL', 'Image','Brand_name','High_level_shape','Price','Size  
        zara_result['Name']=i
        zara_result['Material']=t
        zara_result['Color']=soup.find('span',class_='_colorName').get_text()
        zara_result['URL']=item_link[i]
        zara_result['Image']=image
        zara_result['Brand_name']='Zara'
        zara_result['High_level_shape']=''
        try:
            zara_result['Price']=soup.find('span',class_='main-price').get_text()
        except:
            zara_result['Price']=soup.find('span',class_='line-through').get_text()
        zara_result['Size']=size
        zara_result['Formality']=''
        zara_result['Description']=soup.find('p',class_='description').get_text()
        zara.append(zara_result)
    except:
        pass
#return zara
```
